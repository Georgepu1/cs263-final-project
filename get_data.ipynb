```python
import pandas as pd
tr_dataset = pd.DataFrame({'sentences': sst_data[0]['sentence'], 'labels': sst_data[0]['label']})
val_dataset = pd.DataFrame({'sentences': sst_data[1]['sentence'], 'labels': sst_data[1]['label']})
test_dataset = pd.DataFrame({'sentences': sst_data[2]['sentence'], 'labels': sst_data[2]['label']})
tr_dataset.head()
```

```python
tr_dataset['labels'] = tr_dataset['labels'].apply(lambda x: str(round(x)))
val_dataset['labels'] = val_dataset['labels'].apply(lambda x: str(round(x)))
test_dataset['labels'] = test_dataset['labels'].apply(lambda x: str(round(x)))
tr_dataset.head()
```

```python
print(len(tr_dataset), len(val_dataset), len(test_dataset))
```

```python
cola_tr_dataset = pd.DataFrame({'sentences': [x[0] for x in cola_data[0]], 'labels': [str(x[1]) for x in cola_data[0]]})
cola_val_dataset = pd.DataFrame({'sentences': [x[0] for x in cola_data[1]], 'labels': [str(x[1]) for x in cola_data[1]]})
cola_test_dataset = pd.DataFrame({'sentences': [x[0] for x in cola_data[2]], 'labels': [str(x[1]) for x in cola_data[2]]})
cola_tr_dataset.sample(5)
```

```python
print(len(cola_tr_dataset), len(cola_val_dataset), len(cola_test_dataset))
```

```python
min_train = min(len(cola_tr_dataset), len(tr_dataset))
tr_dataset = tr_dataset[:min_train]
cola_tr_dataset = cola_tr_dataset[:min_train]
```

```python
tr_dataset.to_csv('sst_train_data.csv', index=False)
val_dataset.to_csv('sst_val_data.csv', index=False)
test_dataset.to_csv('sst_test_data.csv', index=False)

cola_tr_dataset.to_csv('cola_train_data.csv', index=False)
cola_val_dataset.to_csv('cola_val_data.csv', index=False)
cola_test_dataset.to_csv('cola_test_data.csv', index=False)
```
